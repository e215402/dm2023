### 演習1.3：ノイズを付与した観測値の準備

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def true_function(x):
    return np.sin(np.pi * x * 0.8) * 10
```

```python
# ランダムに観測点を用意
np.random.seed(0)
x_obs = np.random.uniform(-1, 1, size=20)
```

```python
# 真値を計算
y_true = true_function(x_obs)
```

```python
# ホワイトノイズを生成して付与
noise = np.random.normal(0, 2, size=20) / 2
y_obs = y_true + noise
```

```python
# DataFrameに追加
df = pd.DataFrame({'観測点': x_obs, '真値': y_true, '観測値': y_obs})
```

```python
# 確認用
print(df)
```

```python
#　グラフを描画
x = np.linspace(-1, 1, num=100)
y = true_function(x)
plt.plot(x, y, label='true value')
plt.scatter(x_obs, y_obs, label='Observation point', color='red')
plt.legend()
plt.savefig('ex1.3.png')
plt.show()
```

### 演習1.4：データセットのファイル出力

```python
df.to_csv('data.tsv', index=False)
```
